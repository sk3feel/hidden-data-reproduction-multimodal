```python
# Utilities
import gc, json, os, re, torch, pandas as pd, numpy as np
from pathlib import Path
from collections import defaultdict
from PIL import Image, ImageFile
from transformers import AutoModelForCausalLM, AutoProcessor, Qwen2VLForConditionalGeneration, BitsAndBytesConfig
from peft import PeftModel
from sklearn.metrics import roc_auc_score
from scipy.stats import ttest_ind
from tqdm.auto import tqdm
from qwen_vl_utils import process_vision_info
import matplotlib.pyplot as plt
import seaborn as sns
import comet_ml

import transformers
print(f"transformers version: {transformers.__version__}")
if not transformers.__version__.startswith("4.4"):
    print(f"WARNING: expected 4.4x.x, got {transformers.__version__}. Florence-2 may fail.")


ImageFile.LOAD_TRUNCATED_IMAGES = True
MAX_IMAGE_SIDE = 1024
LOCAL_ARTIFACTS_ROOT = Path("artifacts")
DRIVE_ARTIFACTS_ROOT = DRIVE_ROOT / 'artifacts_v2_controlled'
DRIVE_ARTIFACTS_ROOT.mkdir(parents=True, exist_ok=True)
FINETUNING_DIR = DRIVE_ARTIFACTS_ROOT / "finetuning_generative"
FINETUNING_DIR.mkdir(parents=True, exist_ok=True)
MIA_DIR = DRIVE_ARTIFACTS_ROOT / "privacy_attacks" / "mia"
MIA_DIR.mkdir(parents=True, exist_ok=True)


def save_csv_with_comet(df, path, experiment, table_name=None):
    path = Path(path)
    path.parent.mkdir(parents=True, exist_ok=True)
    df.to_csv(path, index=False)
    csv_string = df.to_csv(index=False)
    experiment.log_asset(str(path), file_name=path.name)
    experiment.log_asset_data(csv_string, file_name=path.name)
    if table_name is not None:
        experiment.log_table(table_name, df)
    return path


def safe_open_image(path, max_side=MAX_IMAGE_SIDE):
    image = Image.open(path).convert("RGB")
    if max(image.size) > max_side:
        image.thumbnail((max_side, max_side), Image.LANCZOS)
    return image


def load_jsonl(path):
    rows = []
    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            if line.strip():
                rows.append(json.loads(line))
    return rows


def enrich_with_field_types(records, manifest_path):
    manifest = []
    with open(manifest_path, "r", encoding="utf-8") as f:
        for line in f:
            if line.strip():
                manifest.append(json.loads(line))
    type_map = {}
    for row in manifest:
        eid = str(row.get("example_id", row.get("question_id", "")))
        type_map[eid] = row.get("coarse_field_type", "OTHER")
    for rec in records:
        rec["coarse_field_type"] = type_map.get(rec["example_id"], "OTHER")
    return records


def stratified_sample(records, n, type_key="coarse_field_type"):
    by_type = defaultdict(list)
    for rec in records:
        by_type[rec[type_key]].append(rec)
    types = sorted(by_type.keys())
    if not types:
        return []
    base = n // len(types)
    remainder = n % len(types)
    target_counts = {ft: base + (1 if idx < remainder else 0) for idx, ft in enumerate(types)}
    result = []
    used = set()
    for ft in types:
        selected = by_type[ft][:target_counts[ft]]
        result.extend(selected)
        used.update(rec["example_id"] for rec in selected)
    if len(result) < n:
        for rec in records:
            if len(result) >= n:
                break
            if rec["example_id"] not in used:
                result.append(rec)
                used.add(rec["example_id"])
    return result[:n]


def make_qwen_user_text(question):
    return f"Answer the question about this document image. Give only the answer value, nothing else.\nQuestion: {question}"


def normalize_text(text):
    return " ".join(str(text or "").strip().lower().split())


def run_qwen_sanity_rows(model, processor, seen_records, unseen_records, n_seen=100, n_unseen=100):
    rows = []
    for split_name, records, limit in [("seen", seen_records, n_seen), ("unseen", unseen_records, n_unseen)]:
        for rec in records[:limit]:
            image = safe_open_image(rec["image_path"])
            messages = [{"role": "user", "content": [{"type": "image", "image": image}, {"type": "text", "text": make_qwen_user_text(rec["question"])}]}]
            text = processor.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
            img_inputs, _ = process_vision_info(messages)
            img_input = img_inputs[0] if isinstance(img_inputs, list) else img_inputs
            inputs = processor(text=[text], images=[img_input], return_tensors="pt", padding=True)
            device = next(model.parameters()).device
            inputs = {k: v.to(device) if isinstance(v, torch.Tensor) else v for k, v in inputs.items()}
            with torch.no_grad():
                gen = model.generate(**inputs, max_new_tokens=50)
            pred = processor.batch_decode(gen[:, inputs["input_ids"].shape[1]:], skip_special_tokens=True)[0].strip()
            rows.append({
                "split": split_name,
                "example_id": rec["example_id"],
                "question": rec["question"],
                "gold": rec["answer"],
                "prediction": pred,
                "exact_match": normalize_text(pred) == normalize_text(rec["answer"]),
            })
    df = pd.DataFrame(rows)
    seen_em = float(df[df["split"] == "seen"]["exact_match"].mean()) if not df[df["split"] == "seen"].empty else 0.0
    unseen_em = float(df[df["split"] == "unseen"]["exact_match"].mean()) if not df[df["split"] == "unseen"].empty else 0.0
    return df, seen_em, unseen_em


def ensure_qwen_sanity_tables(qwen_seen, unseen_200):
    bnb = BitsAndBytesConfig(load_in_4bit=True, bnb_4bit_compute_dtype=torch.bfloat16, bnb_4bit_quant_type="nf4")
    configs = [
        ("qwen2b", "Qwen/Qwen2-VL-2B-Instruct", "checkpoints/qwen2b_v2/epoch_10"),
        ("qwen7b", "Qwen/Qwen2-VL-7B-Instruct", "checkpoints/qwen7b_v2/epoch_10"),
    ]
    for tag, base_model_name, adapter_path in configs:
        output_path = FINETUNING_DIR / f"{tag}_sanity_final.csv"
        if output_path.exists():
            print({"sanity_csv": str(output_path), "status": "exists"})
            continue
        base = Qwen2VLForConditionalGeneration.from_pretrained(base_model_name, quantization_config=bnb, device_map="auto", trust_remote_code=True)
        model = PeftModel.from_pretrained(base, adapter_path)
        model.eval()
        processor = AutoProcessor.from_pretrained(base_model_name, trust_remote_code=True)
        sanity_df, seen_em, unseen_em = run_qwen_sanity_rows(model, processor, qwen_seen, unseen_200, n_seen=100, n_unseen=100)
        output_path.parent.mkdir(parents=True, exist_ok=True)
        sanity_df.to_csv(output_path, index=False)
        print({"sanity_csv": str(output_path), "seen_em": seen_em, "unseen_em": unseen_em})
        del model, base, processor
        cleanup_gpu()


def get_generated_token_ids(gen):
    if not getattr(gen, "scores", None):
        return gen.sequences[:, :0]
    generated_len = len(gen.scores)
    if gen.sequences.shape[1] >= generated_len:
        return gen.sequences[:, -generated_len:]
    return gen.sequences


def get_sequence_log_probs(model, processor, gen, generated_token_ids):
    if generated_token_ids.shape[1] == 0 or not getattr(gen, "scores", None):
        return []
    eos_token_id = processor.tokenizer.eos_token_id
    log_probs = []
    try:
        transition_scores = model.compute_transition_scores(gen.sequences, gen.scores, normalize_logits=True)
        transition_scores = transition_scores[:, -generated_token_ids.shape[1]:]
        for tok, score in zip(generated_token_ids[0].tolist(), transition_scores[0].tolist()):
            tok = int(tok)
            if eos_token_id is not None and tok == eos_token_id:
                break
            if np.isfinite(score):
                log_probs.append(float(score))
    except Exception:
        for i, scores in enumerate(gen.scores):
            if i >= generated_token_ids.shape[1]:
                break
            lsm = torch.nn.functional.log_softmax(scores[0].float(), dim=-1)
            tok = int(generated_token_ids[0, i].item())
            if eos_token_id is not None and tok == eos_token_id:
                break
            value = float(lsm[tok].item())
            if np.isfinite(value):
                log_probs.append(value)
    return log_probs


def summarize_generation_debug(processor, gen, generated_token_ids, log_probs, prompt_len, model_type):
    first_token = None
    if generated_token_ids.shape[1] > 0:
        first_token = int(generated_token_ids[0, 0].item())
    eos_token_id = processor.tokenizer.eos_token_id
    return {
        "model_type": model_type,
        "prompt_len": int(prompt_len),
        "sequence_len": int(gen.sequences.shape[1]),
        "generated_steps": int(len(gen.scores)),
        "generated_token_count": int(generated_token_ids.shape[1]),
        "first_generated_token_id": first_token,
        "first_is_eos": bool(first_token == eos_token_id) if first_token is not None and eos_token_id is not None else False,
        "eos_token_id": int(eos_token_id) if eos_token_id is not None else None,
        "generated_token_ids": [int(x) for x in generated_token_ids[0].tolist()] if generated_token_ids.shape[1] > 0 else [],
        "decoded_generated": processor.batch_decode(generated_token_ids, skip_special_tokens=False)[0] if generated_token_ids.shape[1] > 0 else "",
        "log_probs": [float(x) for x in log_probs],
    }


def compute_confidence_and_loss(model, processor, image, question, gold_answer, model_type="florence2", verbose=False):
    model.eval()
    with torch.no_grad():
        if model_type == "florence2":
            inputs = processor(text=f"<VQA>{question}", images=image, return_tensors="pt").to(model.device)
            labels = processor.tokenizer(gold_answer, return_tensors="pt").input_ids.to(model.device)
            outputs = model(input_ids=inputs["input_ids"], pixel_values=inputs["pixel_values"], labels=labels)
            loss = outputs.loss.item()
            generate_inputs = {"input_ids": inputs["input_ids"], "pixel_values": inputs["pixel_values"]}
            gen = model.generate(**generate_inputs, max_new_tokens=50, output_scores=True, return_dict_in_generate=True)
            generated_token_ids = get_generated_token_ids(gen)
            prompt_len = inputs["input_ids"].shape[1]
            log_probs = get_sequence_log_probs(model, processor, gen, generated_token_ids)
            confidence = sum(log_probs) / max(len(log_probs), 1)
            if verbose:
                print(summarize_generation_debug(processor, gen, generated_token_ids, log_probs, prompt_len, model_type))
        elif model_type == "qwen":
            messages = [{"role": "user", "content": [{"type": "image", "image": image}, {"type": "text", "text": make_qwen_user_text(question)}]}]
            text = processor.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
            img_inputs, _ = process_vision_info(messages)
            img_input = img_inputs[0] if isinstance(img_inputs, list) else img_inputs
            inputs = processor(text=[text], images=[img_input], return_tensors="pt", padding=True)
            device = next(model.parameters()).device
            inputs = {k: v.to(device) if isinstance(v, torch.Tensor) else v for k, v in inputs.items()}
            full_messages = messages + [{"role": "assistant", "content": [{"type": "text", "text": gold_answer}]}]
            full_text = processor.apply_chat_template(full_messages, tokenize=False, add_generation_prompt=False)
            full_inputs = processor(text=[full_text], images=[img_input], return_tensors="pt", padding=True)
            full_inputs = {k: v.to(device) if isinstance(v, torch.Tensor) else v for k, v in full_inputs.items()}
            labels = full_inputs["input_ids"].clone()
            labels[full_inputs["attention_mask"] == 0] = -100
            labels[:, :inputs["input_ids"].shape[1]] = -100
            outputs = model(**full_inputs, labels=labels)
            loss = outputs.loss.item()
            gen = model.generate(**inputs, max_new_tokens=50, output_scores=True, return_dict_in_generate=True)
            generated_token_ids = get_generated_token_ids(gen)
            prompt_len = inputs["input_ids"].shape[1]
            log_probs = get_sequence_log_probs(model, processor, gen, generated_token_ids)
            confidence = sum(log_probs) / max(len(log_probs), 1)
            if verbose:
                print(summarize_generation_debug(processor, gen, generated_token_ids, log_probs, prompt_len, model_type))
        else:
            raise ValueError(f"Unknown model_type: {model_type}")
    if not np.isfinite(confidence):
        confidence = -100.0
    if not np.isfinite(loss):
        loss = 100.0
    return confidence, loss


def run_mia(model, processor, seen, unseen, model_type):
    results = []
    for rec in tqdm(seen, desc="MIA seen"):
        img = safe_open_image(rec["image_path"])
        c, l = compute_confidence_and_loss(model, processor, img, rec["question"], rec["answer"], model_type)
        results.append({"example_id": rec["example_id"], "split": "seen", "field_type": rec["coarse_field_type"], "confidence": c, "loss": l})
    for rec in tqdm(unseen, desc="MIA unseen"):
        img = safe_open_image(rec["image_path"])
        c, l = compute_confidence_and_loss(model, processor, img, rec["question"], rec["answer"], model_type)
        results.append({"example_id": rec["example_id"], "split": "unseen", "field_type": rec["coarse_field_type"], "confidence": c, "loss": l})
    return pd.DataFrame(results)


def sanitize_mia_dataframe(df):
    clean = df.copy()
    clean["confidence"] = pd.to_numeric(clean["confidence"], errors="coerce")
    clean["loss"] = pd.to_numeric(clean["loss"], errors="coerce")
    clean = clean.replace([np.inf, -np.inf], np.nan)
    dropped = int(clean[["confidence", "loss"]].isna().any(axis=1).sum())
    clean = clean.dropna(subset=["confidence", "loss"]).reset_index(drop=True)
    return clean, dropped


def compute_mia_metrics(df):
    df = df.copy()
    df = df[np.isfinite(df["confidence"]) & np.isfinite(df["loss"])]
    if len(df) < 10:
        return {
            "auc_confidence": 0.5,
            "auc_loss": 0.5,
            "t_stat": 0,
            "p_value": 1,
            "mean_seen_conf": 0,
            "std_seen_conf": 0,
            "mean_unseen_conf": 0,
            "std_unseen_conf": 0,
            "mean_seen_loss": 0,
            "mean_unseen_loss": 0,
        }
    labels = (df["split"] == "seen").astype(int)
    sc = df[df["split"] == "seen"]["confidence"]
    uc = df[df["split"] == "unseen"]["confidence"]
    sl = df[df["split"] == "seen"]["loss"]
    ul = df[df["split"] == "unseen"]["loss"]
    t_stat, p_value = ttest_ind(sc, uc, equal_var=False)
    return {
        "auc_confidence": roc_auc_score(labels, df["confidence"]),
        "auc_loss": roc_auc_score(labels, -df["loss"]),
        "t_stat": t_stat,
        "p_value": p_value,
        "mean_seen_conf": sc.mean(),
        "std_seen_conf": sc.std(),
        "mean_unseen_conf": uc.mean(),
        "std_unseen_conf": uc.std(),
        "mean_seen_loss": sl.mean(),
        "mean_unseen_loss": ul.mean(),
    }


def plot_histogram(df, title):
    clean, dropped = sanitize_mia_dataframe(df)
    fig, ax = plt.subplots(figsize=(8, 5))
    ax.hist(clean[clean["split"] == "seen"]["confidence"], bins=30, alpha=0.6, label="seen", color="red")
    ax.hist(clean[clean["split"] == "unseen"]["confidence"], bins=30, alpha=0.6, label="unseen", color="blue")
    ax.set_xlabel("??????????? ?????? (??????? log-prob)")
    ax.set_ylabel("?????????? ????????")
    suffix = f" | ?????????: {dropped}" if dropped else ""
    ax.set_title(title + suffix)
    ax.legend()
    plt.tight_layout()
    return fig


def cleanup_gpu():
    gc.collect()
    torch.cuda.empty_cache()
```

```python
# Data preparation
local_common_train_ids_path = Path('artifacts/finetuning_generative/common_train_200_ids.csv')
common_train_ids_path = FINETUNING_DIR / 'common_train_200_ids.csv'
common_unseen_ids_path = FINETUNING_DIR / 'common_unseen_200_ids.csv'

all_train = load_jsonl('artifacts/docqa_recovery/benchmark_train/manifest.jsonl')
all_train_map = {str(r.get('example_id', r.get('question_id', ''))): r for r in all_train}
all_val = load_jsonl('artifacts/docqa_recovery/benchmark/manifest.jsonl')
all_val_map = {str(r.get('example_id', r.get('question_id', ''))): r for r in all_val}

def manifest_to_record(row, split, image_root):
    eid = str(row.get('example_id', row.get('question_id', '')))
    image_name = str(row.get('image_path', '')).replace('\\', '/').split('/')[-1]
    answer_value = row['answers'][0] if isinstance(row.get('answers'), list) and row.get('answers') else row.get('answer', '')
    return {'example_id': eid, 'image_path': f'{image_root}/{image_name}', 'question': row['question'], 'answer': answer_value, 'coarse_field_type': row.get('coarse_field_type', 'OTHER')}

train_records_full = [manifest_to_record(r, 'seen', 'artifacts/docqa_recovery/benchmark_train/images/original') for r in all_train]
val_records_full = [manifest_to_record(r, 'unseen', 'artifacts/docqa_recovery/benchmark/images/original') for r in all_val]

if common_train_ids_path.exists():
    common_train_ids = pd.read_csv(common_train_ids_path)['example_id'].astype(str).tolist()
elif local_common_train_ids_path.exists():
    common_train_ids = pd.read_csv(local_common_train_ids_path)['example_id'].astype(str).tolist()
    pd.DataFrame({'example_id': common_train_ids}).to_csv(common_train_ids_path, index=False)
else:
    common_train_ids = [r['example_id'] for r in stratified_sample(train_records_full, 200)]
    pd.DataFrame({'example_id': common_train_ids}).to_csv(common_train_ids_path, index=False)

if common_unseen_ids_path.exists():
    common_unseen_ids = pd.read_csv(common_unseen_ids_path)['example_id'].astype(str).tolist()
else:
    common_unseen_ids = [r['example_id'] for r in stratified_sample(val_records_full, 200)]
    pd.DataFrame({'example_id': common_unseen_ids}).to_csv(common_unseen_ids_path, index=False)

train_map = {r['example_id']: r for r in train_records_full}
val_map = {r['example_id']: r for r in val_records_full}
common_seen = [train_map[eid] for eid in common_train_ids if eid in train_map]
unseen_200 = [val_map[eid] for eid in common_unseen_ids if eid in val_map]
qwen_seen = common_seen
florence_seen = common_seen
print(f'Florence seen: {len(florence_seen)}, Qwen seen: {len(qwen_seen)}, Unseen: {len(unseen_200)}')
```

```python
# Qwen confidence on one example
rec = qwen_seen[0]
print("example_id:", rec["example_id"])
print("question:", rec["question"])
print("answer:", rec["answer"])

bnb = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.bfloat16,
    bnb_4bit_quant_type="nf4",
)

model = Qwen2VLForConditionalGeneration.from_pretrained(
    "Qwen/Qwen2-VL-2B-Instruct",
    quantization_config=bnb,
    device_map="auto",
    trust_remote_code=True,
)
model.eval()

processor = AutoProcessor.from_pretrained(
    "Qwen/Qwen2-VL-2B-Instruct",
    trust_remote_code=True,
)

image = safe_open_image(rec["image_path"])
messages = [{
    "role": "user",
    "content": [
        {"type": "image", "image": image},
        {"type": "text", "text": make_qwen_user_text(rec["question"])},
    ],
}]

text = processor.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
img_inputs, _ = process_vision_info(messages)
img_input = img_inputs[0] if isinstance(img_inputs, list) else img_inputs

inputs = processor(text=[text], images=[img_input], return_tensors="pt", padding=True)
device = next(model.parameters()).device
inputs = {k: v.to(device) if isinstance(v, torch.Tensor) else v for k, v in inputs.items()}

print("tokenizer eos_token_id:", processor.tokenizer.eos_token_id)
print("tokenizer eos_token:", processor.tokenizer.eos_token)
print("input_ids shape:", tuple(inputs["input_ids"].shape))
print("prompt_len:", inputs["input_ids"].shape[1])

with torch.no_grad():
    gen = model.generate(
        **inputs,
        max_new_tokens=50,
        output_scores=True,
        return_dict_in_generate=True,
    )

print("gen.sequences shape:", tuple(gen.sequences.shape))
print("len(gen.scores):", len(gen.scores))

prompt_len = inputs["input_ids"].shape[1]
seq_len = gen.sequences.shape[1]

if seq_len > prompt_len:
    new_tokens_old_logic = gen.sequences[0, prompt_len:]
else:
    new_tokens_old_logic = torch.tensor([], device=gen.sequences.device, dtype=gen.sequences.dtype)

generated_len = len(gen.scores)
if generated_len > 0 and seq_len >= generated_len:
    new_tokens_tail = gen.sequences[0, -generated_len:]
else:
    new_tokens_tail = torch.tensor([], device=gen.sequences.device, dtype=gen.sequences.dtype)

print("new_tokens_old_logic len:", int(new_tokens_old_logic.shape[0]))
print("new_tokens_tail len:", int(new_tokens_tail.shape[0]))

if new_tokens_old_logic.shape[0] > 0:
    print("first old_logic token id:", int(new_tokens_old_logic[0].item()))
    print("first old_logic is eos:", int(new_tokens_old_logic[0].item()) == processor.tokenizer.eos_token_id)
else:
    print("old_logic produced 0 new tokens")

if new_tokens_tail.shape[0] > 0:
    print("first tail token id:", int(new_tokens_tail[0].item()))
    print("first tail is eos:", int(new_tokens_tail[0].item()) == processor.tokenizer.eos_token_id)
else:
    print("tail extraction produced 0 new tokens")

print("decoded old_logic:", processor.batch_decode(new_tokens_old_logic.unsqueeze(0), skip_special_tokens=False)[0] if new_tokens_old_logic.shape[0] > 0 else "")
print("decoded tail:", processor.batch_decode(new_tokens_tail.unsqueeze(0), skip_special_tokens=False)[0] if new_tokens_tail.shape[0] > 0 else "")

log_probs_old_logic = []
if len(gen.scores) > 0:
    for i, scores in enumerate(gen.scores):
        lsm = torch.nn.functional.log_softmax(scores[0], dim=-1)

        if i + prompt_len < seq_len:
            tok_old = int(gen.sequences[0, i + prompt_len].item())
            if tok_old == processor.tokenizer.eos_token_id:
                break
            log_probs_old_logic.append(lsm[tok_old].item())

print("log_probs_old_logic:", log_probs_old_logic[:10])
print("confidence_old_logic:", sum(log_probs_old_logic) / max(len(log_probs_old_logic), 1))

log_probs_tail = []
if len(gen.scores) > 0 and new_tokens_tail.shape[0] > 0:
    for i, scores in enumerate(gen.scores):
        if i >= new_tokens_tail.shape[0]:
            break
        lsm = torch.nn.functional.log_softmax(scores[0], dim=-1)
        tok = int(new_tokens_tail[i].item())
        if tok == processor.tokenizer.eos_token_id:
            break
        log_probs_tail.append(lsm[tok].item())

print("log_probs_tail:", log_probs_tail[:10])
print("confidence_tail:", sum(log_probs_tail) / max(len(log_probs_tail), 1))

c, l = compute_confidence_and_loss(
    model,
    processor,
    image,
    rec["question"],
    rec["answer"],
    model_type="qwen",
    verbose=True,
)
print("compute_confidence_and_loss ->", {"confidence": c, "loss": l})
```

```python
# AUC by field type
summary_experiment = comet_ml.Experiment(api_key=os.environ["COMET_API_KEY"], workspace="scfeel", project_name="qwen3-1")
summary_experiment.set_name("mia-summary-v2")

baseline_summary_df = pd.read_csv(MIA_DIR / "baseline_mia_summary.csv")
finetuned_summary_df = pd.read_csv(MIA_DIR / "finetuned_mia_summary.csv")
summary_experiment.log_table("baseline_mia_summary.csv", baseline_summary_df)
summary_experiment.log_table("finetuned_mia_summary.csv", finetuned_summary_df)

fine_tuned_files = {
    "florence2_epoch30": MIA_DIR / "florence2_epoch10.csv",
    "qwen2b_epoch20": MIA_DIR / "qwen2b_epoch10.csv",
    "qwen7b_epoch10": MIA_DIR / "qwen7b_epoch10.csv",
}
rows = []
for model_tag, csv_path in fine_tuned_files.items():
    df = pd.read_csv(csv_path)
    for field_type, group in df.groupby("field_type"):
        labels = (group["split"] == "seen").astype(int)
        if labels.nunique() < 2:
            continue
        rows.append({
            "model": model_tag,
            "field_type": field_type,
            "auc_confidence": roc_auc_score(labels, group["confidence"]),
            "auc_loss": roc_auc_score(labels, -group["loss"]),
            "n_examples": len(group),
        })
field_auc_df = pd.DataFrame(rows).sort_values(["model", "field_type"]).reset_index(drop=True)
save_csv_with_comet(field_auc_df, MIA_DIR / "field_type_auc.csv", summary_experiment, table_name="field_type_auc.csv")

fig, ax = plt.subplots(figsize=(10, 4))
heatmap_df = field_auc_df.pivot(index="model", columns="field_type", values="auc_confidence")
sns.heatmap(heatmap_df, annot=True, fmt=".3f", cmap="YlOrRd", ax=ax)
ax.set_title("MIA AUC-ROC по типам полей")
ax.set_xlabel("Тип поля")
ax.set_ylabel("Модель")
plt.tight_layout()
summary_experiment.log_figure("field_type_auc_heatmap", fig)
plt.close(fig)

mia_bar_rows = []
for _, row in baseline_summary_df.iterrows():
    mia_bar_rows.append({"tag": row["tag"], "stage": "baseline", "auc_confidence": row["auc_confidence"], "auc_loss": row["auc_loss"]})
for _, row in finetuned_summary_df.iterrows():
    mia_bar_rows.append({"tag": row["tag"], "stage": "fine-tuned", "auc_confidence": row["auc_confidence"], "auc_loss": row["auc_loss"]})
mia_bar_df = pd.DataFrame(mia_bar_rows)
save_csv_with_comet(mia_bar_df, MIA_DIR / "mia_auc_bar_summary.csv", summary_experiment, table_name="mia_auc_bar_summary.csv")

plot_df = mia_bar_df.melt(id_vars=["tag", "stage"], value_vars=["auc_confidence", "auc_loss"], var_name="metric", value_name="auc")
fig, axes = plt.subplots(1, 2, figsize=(12, 4), sharey=True)
for ax, metric in zip(axes, ["auc_confidence", "auc_loss"]):
    metric_df = plot_df[plot_df["metric"] == metric]
    sns.barplot(data=metric_df, x="tag", y="auc", hue="stage", ax=ax, palette="Set2")
    ax.set_title("AUC-ROC по " + ("confidence" if metric == "auc_confidence" else "loss"))
    ax.set_xlabel("Модель")
    ax.set_ylabel("AUC-ROC")
    ax.axhline(0.5, linestyle="--", color="gray", linewidth=1)
plt.tight_layout()
summary_experiment.log_figure("mia_auc_bar_baseline_vs_finetuned", fig)
plt.close(fig)

display(field_auc_df)
display(mia_bar_df)
summary_experiment.end()
```
