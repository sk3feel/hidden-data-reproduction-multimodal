# 01. Inspect DocVQA

Быстрый просмотр датасета DocVQA из Hugging Face (`pixparse/docvqa-single-page-questions`).

```python
import os
print(os.getenv("HF_TOKEN") is not None)
```

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path('..').resolve()
SRC_DIR = PROJECT_ROOT / 'src'
if str(SRC_DIR) not in sys.path:
    sys.path.append(str(SRC_DIR))

from load_data import load_docvqa, show_example
```

```python
examples = load_docvqa(split='validation', limit=3)
print('Loaded examples:', len(examples))
print('Keys:', list(examples[0].keys()))
print('Type(image):', type(examples[0].get('image')))
print('Type(ocr_results):', type(examples[0].get('ocr_results')))
```

```python
for i in range(min(5, len(examples))):
    print('=' * 80)
    print('Example', i)
    show_example(examples[i])
```

```python
from PIL import Image

out_dir = PROJECT_ROOT / 'outputs' / 'examples'
out_dir.mkdir(parents=True, exist_ok=True)

for i, ex in enumerate(examples[:5]):
    img = ex.get('image')
    if isinstance(img, Image.Image):
        pil = img
    elif isinstance(img, dict) and img.get('bytes') is not None:
        from io import BytesIO
        pil = Image.open(BytesIO(img['bytes'])).convert('RGB')
    elif isinstance(img, dict) and img.get('path'):
        pil = Image.open(img['path']).convert('RGB')
    else:
        continue
    save_path = out_dir / f'raw_example_{i}.png'
    pil.save(save_path)
    print('Saved:', save_path)
```
